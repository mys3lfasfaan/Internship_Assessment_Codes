```python
def find_empty_cells(df):
    empty_cells = df.isnull()
    positions = [(i, j) for i in range(df.shape[0]) for j in range(df.shape[1]) if empty_cells.iloc[i, j]]
    return positions

# Test the function
print(find_empty_cells(df))
```

```python
float_cols = df.columns[2:]
for col in float_cols:
    df[col] = df[col].str.split(' ', expand=True)[0]
    
# Convert columns to float values  
df[float_cols] = df[float_cols].astype(float)
df[float_cols] = df[float_cols] * 100000 # Convert lacs to actual number
```
